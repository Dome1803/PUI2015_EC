```python
import pandas as pd
import sys
import os
import numpy 
import pylab as pl

import json
import os
%pylab inline
```

```python
homeless=pd.read_csv('homeless_charter_2014.csv')
homeless.head()
```

```python
scores=pd.read_csv('scores_charter_2014.csv')
scores.head()
```

```python
income=pd.read_csv('income_zip_2014.csv')
income.head()
```

```python
zipcode=pd.read_csv('school_beds_zipcode.csv')
zipcode.head()
```

```python
homeless_zip=pd.merge(homeless, zipcode, left_on='BEDS', right_on='BEDS Number', how='left')
homeless_zip.head()
```

```python
income['sub10pct'] = income['sub10k']/income['total']*100
income['10-15pct'] = income['10-15']/income['total']*100
income['15-20pct'] = income['15-20']/income['total']*100
income['20-25pct'] = income['20-25']/income['total']*100
income['25-30pct'] = income['25-30']/income['total']*100
poverty = income[['zip', 'sub10pct','10-15pct','15-20pct','20-25pct','25-30pct']]
poverty.head()
```

```python
school_poverty=pd.merge(homeless_zip, poverty, left_on='Zip', right_on='zip', how='left')
school_poverty=school_poverty.drop(['Zip','BEDS Number'],1)
school_poverty.head()
```
